```python
def cast(x):
    if x == 0 :
        return 0
    elif x > 0 and x <= 0.1:
        return 0.1
    elif x > 0.1 and x <= 0.2:
        return 0.2
    elif x > 0.2 and x <= 0.3:
        return 0.3
    elif x > 0.3 and x <= 0.4:
        return 0.4
    elif x > 0.4 and x <= 0.5:
        return 0.5
    elif x > 0.5 and x <= 0.6:
        return 0.6
    elif x > 0.6 and x <= 0.7:
        return 0.7
    elif x > 0.7 and x <= 0.8:
        return 0.8
    elif x > 0.8 and x <= 0.9:
        return 0.9
    elif x > 0.9 and x < 1:
        return 1
```

```python
df['ShortTermScore'] = df['ShortTermScore'].apply(cast)
df['LongTermScore'] = df['LongTermScore'].apply(cast)
```

```python
by_shrtscr = (
    df
    .query('RequestedPeriod <= 40')
    .groupby('ShortTermScore')
    .agg({'orderid':'count',
          'IsBlocked':'sum',
          'ManualApprove':'sum',
          'IsManual':'sum',
          'AutoDecision':'sum',
          'IsLoan':'sum'
          })
)
by_shrtscr

# автоматическое одобрение возможно при ShortTermScore от 0.6 
```

```python
by_longtscr = (
    df
    .query('RequestedPeriod > 40')
    .groupby('LongTermScore')
    .agg({'orderid':'count',
          'IsBlocked':'sum',
          'ManualApprove':'sum',
          'IsManual':'sum',
          'AutoDecision':'sum',
          'IsLoan':'sum'
          })
)
by_longtscr

# автоматическое одобрение возможно при LongTermScore от 0.3 и при его отсутствии
```

```python
by_longtscr.orderid.sum() + by_shrtscr.orderid.sum()
```

### Данные по дням недели

```python
# по дням недели
df['dow'] = df['OrderDate'].apply(lambda x: x.weekday())
df_by_day_of_the_week = (
                        df
                        .groupby('dow')
                        .agg({'orderid':'nunique', 'RequestedAmount':['sum','min','max','mean'],
                              'IsLoan':'sum',
                              'LoanSum':['sum','max','mean'],
                              'AutoDecision': 'sum',
                              'ManualApprove': 'sum',
                              'IsBlocked':'sum'})
                        .reset_index()
                    )
df_by_day_of_the_week.columns = ['weekday', 'orders_uniq', 'orders_sum', 'min_order', 'max_order', 'mean_order', 'loans_given', 'loans_sum', 'loan_max', 'loan_mean', 'auto', 'manually', 'blocked']
df_by_day_of_the_week
```

```python
df_by_day_of_the_week['orders_uniq'].value_counts(normalize=True)
```

```python
df_by_day_of_the_week = df_by_day_of_the_week[{'orders_uniq', 'loans_given', 'auto', 'manually', 'blocked'}]
df_by_day_of_the_week = (df_by_day_of_the_week / df_by_day_of_the_week.sum()).round(4)

sns.set(font_scale=1.4)
sns.heatmap(data=df_by_day_of_the_week, annot=True, annot_kws={"fontsize":20},
            fmt='.2%', cmap= 'coolwarm', linewidths=0.5, linecolor='black')
```

```python
# по часам внутри дня
df['hour'] = pd.to_datetime(df['OrderDate'], format='%H:%M:%S').dt.hour
```

```python
df_by_hours = (
    df
    .groupby('hour')
    .agg({'orderid':'nunique', 'RequestedAmount':['sum','min','max','mean'],
          'IsLoan':'sum',
          'LoanSum':['sum','max','mean'],
          'AutoDecision': 'sum',
          'ManualApprove': 'sum',
          'IsBlocked':'sum'})
    )
df_by_hours.columns = ['orders_uniq', 'orders_sum', 'min_order', 'max_order', 'mean_order', 'loans_given', 'loans_sum', 'loan_max', 'loan_mean', 'auto', 'manually', 'blocked']
df_by_hours
```

```python
df_by_hours = df_by_hours[{'orders_uniq', 'loans_given', 'auto', 'manually', 'blocked'}]
df_by_hours = (df_by_hours / df_by_hours.sum()).round(4)

sns.set(font_scale=1.4)
sns.heatmap(data=df_by_hours, annot=True, annot_kws={"fontsize":20},
            fmt='.2%', cmap= 'coolwarm', linewidths=0.5, linecolor='black')
```
